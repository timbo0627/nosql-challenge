## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print("Collections in the DB:")
pprint(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene":20}
result = establishments.find(query)

# Use count_documents to display the number of documents in the result
num_documents = establishments.count_documents(query)
print("Number of documents:", num_documents)

# Display the first document in the results using pprint
for document in result:
    pprint(document)
    break  # Exit after printing the first document
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
num_rows = len(df)
num_rows

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityCode":'530', #found that LocalAuthorityCode 530 is London's code
    "RatingValue":{"$gte": 4}
}
result = establishments.find(query)
num_establishments = establishments.count_documents(query)

# Use count_documents to display the number of documents in the result
print("Number of places in London with rating greater than or equal to 4: ", num_establishments)

# Display the first document in the results using pprint
for document in result:
    pprint(document)
    break  # Exit after printing the first document
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
num_rows = len(df)
num_rows

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142  # Latitude of Penang Flavours
longitude = 0.08384  # Longitude of Penang Flavours

query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}
sort = [("scores.Hygiene", 1)]  # 1 for ascending order
limit = 5

result = establishments.find(query).sort(sort).limit(limit)

# Print the results
for index, document in enumerate(result, 1):
    print(f"Establishment {index}:")
    pprint(document)
    print()
```

```python
# Convert result to Pandas DataFrame
result = establishments.find(query).sort(sort).limit(limit)

df = pd.DataFrame(result)
df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:

pipeline = [
    # Match establishments with a hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},
    # Group by Local Authority
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    # Sort from highest to lowest count
    {"$sort": {"count": -1}}
]
result = establishments.aggregate(pipeline)
result_list = list(result)
num_documents = len(result_list)

# Print the number of documents in the result
print("Number of documents in the result:", num_documents)

# Print the first 10 results
for document in result_list[:10]:
    print(document)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result_list)

# Display the number of rows in the DataFrame
num_rows = len(df)
print("Number of rows in the DataFrame:", num_rows)

# Display the first 10 rows of the DataFrame
df.head(10)
```
